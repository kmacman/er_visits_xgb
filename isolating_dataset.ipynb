# Isolating single US hospital from larger dataset
Original data available at: https://dataverse.harvard.edu/dataset.xhtml;jsessionid=14a51733a6d8f4ce61701457b27c?persistentId=doi:10.7910/DVN/QHPZOX

```python
df = pd.read_excel('DataForRegressions_March2018.xlsx')
```

```python
df.describe()
```

```python
df['country'].value_counts()
```

```python
#filtering for US data
df = df[df['country'] == 'usa']
```

```python
df['city'].value_counts()
```

```python
df['hospital'].value_counts()
```

```python
#drop city and country columns
df = df.drop(columns=['city', 'country'])
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
```
